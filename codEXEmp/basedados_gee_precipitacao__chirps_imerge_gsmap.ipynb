# Inicializando GEE

```python
# importa bibliotecas
import ee, geemap

# inicializando
geemap.ee_initialize(project='d2021028876')

import pandas as pd
```

```python
# verificando a versão instalada do Earth Engine
print(ee.__version__)
```

# [CHIRPS](https://developers.google.com/earth-engine/datasets/catalog/UCSB-CHG_CHIRPS_DAILY?hl=pt-br): `DESDE 1981` - `DIÁRIO` - `5566m` - `2 MESES DE ATRASO`

```python
# carregando os dados
chirps = ee.ImageCollection('UCSB-CHG/CHIRPS/DAILY') \
           .filterDate('2025-02-01', '2025-06-02') \
           .select('precipitation')

# qte de imagens
print('\n---------------------------------\nQuantidade de imagens=', chirps.size().getInfo())

# imagens
print('---------------------------------\nImagens=')
display(chirps)

# data das imagens
print('---------------------------------\nDatas das Imagens=')
chirps = chirps.map(lambda img: img.set( {"DATE": ee.Date(img.get("system:time_start")).format("YYYY-MM-dd hh:mm")}))
agrega = (chirps.aggregate_array("DATE").getInfo())
display(pd.DataFrame(agrega))
print('---------------------------------')
```

# [GSMAP](https://developers.google.com/earth-engine/datasets/catalog/JAXA_GPM_L3_GSMaP_v8_operational?hl=pt-br): `DESDE 1998` - `A CADA 1 HORA` - `11132m` - `2 DIAS DE ATRASO`
